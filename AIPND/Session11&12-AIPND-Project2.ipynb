<u>**Step 2.3.1 Data Transforms (`data_transforms`):**</u>

It is optional to apply transformations to `valid` data. Possible transformations are explained below with an indication to whether they apply to `train`, `test`, or both data.

   - **Random Rotation:** [`train`]
     - **Purpose:** Introduces variability by randomly rotating the image by a maximum of 30 degrees.
     - **Benefit:** Augments the dataset, helping the model generalize better to rotated versions of the training images.

   - **Random Resized Crop (to 224x224):** [`train`]
     - **Purpose:** Randomly crops and resizes the image to 224x224 pixels.
     - **Benefit:** Provides robustness to variations in object size and location, enhancing the model's ability to recognize objects in different contexts.

   - **Random Horizontal Flip:** [`train`]
     - **Purpose:** Randomly flips the image horizontally.
     - **Benefit:** Augments the dataset with horizontally flipped versions, improving the model's ability to handle images with varying orientations.

   - **To Tensor:** [`both`]
     - **Purpose:** Converts the image to a PyTorch tensor.
     - **Benefit:** Transforms the image into a format suitable for input to a neural network, enabling efficient computation during training.

   - **Normalize:** [`both`]
     - **Purpose:** Normalizes the pixel values of the image.
     - **Benefit:** Standardizes the input data, making it centered around zero with a standard deviation of 1. This aids in stable and faster convergence during model training.

   - **Resize (to 255):** [`test`]
     - **Purpose:** Resizes the image to a fixed size of 255 pixels.
     - **Benefit:** Ensures consistency in the input size for testing, facilitating uniform processing during model evaluation.

   - **Center Crop (to 224x224):** [`test`]
     - **Purpose:** Crops the center of the image to 224x224 pixels.
     - **Benefit:** Similar to the training set, provides a standardized input size for testing, ensuring compatibility with the model architecture.

**A Note on Normalization**

**Why Normalize Images:**
Normalization of images is crucial for standardizing pixel values, ensuring that they fall within a consistent and manageable range. This process aids in stabilizing and expediting the convergence of neural network models during training.

**Effect of Normalization on Pixel Value Range:**
Normalization scales the pixel values, typically bringing them to a range centered around zero with a standard deviation of 1. This standardized range prevents numerical instability during computations and enables more efficient weight updates in the training process.

**Normalization Values `[0.485, 0.456, 0.406], [0.229, 0.224, 0.225]`:**
These specific values are derived from the mean and standard deviation of the ImageNet dataset, which is a widely used dataset for training deep learning models. Using these values helps align the input data distribution with what the pre-trained models have been trained on. This alignment is essential for effective transfer learning, where a pre-trained model is fine-tuned on a new dataset, ensuring the model adapts smoothly to the new data distribution.

**To Summarise:**
The specified data transformations, applied during both training and testing, serve the purpose of augmenting the training dataset, ensuring standardized input sizes, and normalizing pixel values. These transformations contribute to improved model generalization, robustness, and consistent performance during both training and evaluation phases.

<u>**Step 3.3 Train the Classifier:**</u>

In the training phase, the classifier for the flower recognition model is fine-tuned over five epochs. Each epoch involves iterating through batches of training data, updating the model parameters through backpropagation and optimization. Validation is performed at regular intervals to monitor the model's performance on unseen data. Key statistics, including training and validation loss, as well as accuracy, are printed to provide insights into the model's progress. Proper state management ensures correct behavior of layers like dropout during training and evaluation. The process aims to optimize the classifier for accurate flower classification, leveraging transfer learning from a pre-trained model while tailoring it to the specific task at hand.

**Here are the details:**

1. **Initialization:**
   - Initialize the number of training epochs (a short value between 5 and 10 should do in general).
   - Initialize counters for steps, running loss, and specify the validation frequency (could be between 5 to 20 in general).

2. **Epoch Loop:**
   - Iterate over the specified number of epochs.

3. **Training Loop:**
   - Iterate over batches of training data.
   - Increment the step counter.
   - Move input data and labels to the specified device (e.g., GPU).

4. **Zero Gradient, Forward Pass, Backward Pass, and Optimization:**
   - Zero out gradients to avoid accumulation.
   - Perform a forward pass to obtain log probabilities.
   - Calculate the loss using Negative Log Likelihood Loss.
   - Perform a backward pass to compute gradients.
   - Update model parameters using the optimizer.

5. **Running Loss Accumulation:**
   - Accumulate the running loss for the current batch.

6. **Validation Checkpoint:**
   - Check if it's time to perform validation based on the specified frequency.
   - Initialize variables for validation loss and accuracy.
   - Set the model to evaluation mode.
   - Disable gradient computation during validation.
   - Iterate over batches of validation data.
   - Calculate validation loss and accuracy.
   - Print training and validation statistics.

7. **Model State Management:**
   - Reset the running loss for the next set of training steps.
   - Set the model back to training mode.

**Benefits:**
   - **Iterative Training:** Incremental updates to model parameters using batches of training data.
   - **Validation Monitoring:** Periodic validation checks to monitor model performance on unseen data.
   - **Printed Statistics:** Displaying key statistics provides insights into model training progress.
   - **Model State Management:** Ensures correct behavior of layers like dropout during training and evaluation.

<FONT COLOR="RED">**Here is the sample pseudocode for step 3.3**

```
Set epochs to 5
Initialize steps to 0, running_loss to 0, and valid_freq to 20

For each epoch:
    For each batch in training data:
        Increment steps
        Move input data and labels to device

        Clear gradient
        Perform forward pass to get log probabilities
        Calculate loss using specified criterion
        Perform backward pass to compute gradients
        Update model parameters using optimizer
        
        Accumulate running loss

        If steps is a multiple of valid_freq:
            Initialize valid_loss to 0 and accuracy to 0
            Set model to evaluation mode
            For each batch in validation data:
                Move validation data to device
                Perform forward pass for validation
                Calculate validation loss for the batch
                Accumulate validation loss
                Calculate exponentials of log probabilities
                Check equality between true labels and predicted labels
                Accumulate accuracy

            Print epoch, training loss, validation loss, and validation accuracy
            Reset running loss
            Set model back to training mode
```
