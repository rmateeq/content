## **Linear Combinations**

A **linear combination** is a vector operation that involves multiplying vectors by scalars and adding them together **[scalar multiplication and vector addition]**.

**Formally,** a linear combination of vectors v1, v2, ..., vn is:

- c1v1 + c2v2 + ... + cn*vn

Where c1, c2, ..., cn are scalars.

Intuitively, it involves:

- Taking vectors
- Scaling them by scalars (stretching/shrinking and possibly change in direction)
- Adding the resulting scaled vectors
- This produces a new vector that lies in the span of the original vectors.

**Some examples:**

- 3v1 + 2v2
- 1.5v1 - 2.3v2 + 0.7v3

### **Isn't it like linear regression??**

**Properties:**

- Order of vectors doesn't matter
- Scaling and adding is distributive
- Any vector in the span can be generated

**Linear combinations are fundamental vector operations that enable generating new vectors from existing ones. They allow expressing vectors in terms of basis vectors as well.**

So, we have **two new terms** up our sleeves:
- **span**, and
- **basis vectors**

### **Linear Combinations Illustrated**

```python
import numpy as np
import matplotlib.pyplot as plt
# Define two vectors
v1 = np.array([2, 1])
v2 = np.array([-1, 3])

# Create a plot
plt.figure(figsize=(8, 8))

# Plotting the vectors
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='b', label='v1')
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='g', label='v2')

# Creating linear combinations
lc1 = 2 * v1 - 3 * v2
lc2 = -1 * v1 + 2 * v2

# Plotting linear combinations
plt.quiver(0, 0, lc1[0], lc1[1], angles='xy', scale_units='xy', scale=1, color='r', label='2v1 - 3v2')
plt.quiver(0, 0, lc2[0], lc2[1], angles='xy', scale_units='xy', scale=1, color='m', label='-v1 + 2v2')

# Set plot limits
plt.xlim(-7, 7)
plt.ylim(-7, 7)

# Labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Combinations of Vectors')
plt.axhline(0, color='gray', linewidth=0.8)
plt.axvline(0, color='gray', linewidth=0.8)

# Grid and legend
plt.grid(True)
plt.legend()

plt.show()
```

- In both cases the new vectors are scaled and added to result in new vectors.
- These new vectors are the in linear combination of the actual vectors.

### **Ready to Define Span!**

The **span** of a vector (or a set of vectors) is the **set of all possible linear combinations** that can be formed using scalar multiples and addition of those vectors. In other words, it represents all the points that can be reached by scaling and adding the vectors together. The **span is a subspace of the vector space** containing the original vectors.

**What is a vector space though?**

A vector space is a collection of objects called vectors that behave in a well-defined, consistent way when you add them together or multiply them by numbers.

For example, the 2D plane consists of all possible (x, y) coordinate vectors. This forms a vector space because:

- You can add any two (x, y) vectors by adding their x and y components: (x1, y1) + (x2, y2) = (x1+x2, y1+y2)
- You can multiply any (x, y) vector by a number to scale it: 3*(x, y) = (3x, 3y)
- These operations don't take you outside the set of (x, y) vectors. The results are still vectors in the 2D plane.
- The operations follow consistent, reliable rules like associativity and distributivity.

### **Let's Visualize Span!**

```python
# Define the vector
v = np.array([2, 1])

# Define scalar multiples
s1 = 2
s2 = -1

# Calculate scaled vectors
scaled_v1 = s1 * v
scaled_v2 = s2 * v

# Create a plot
plt.figure(figsize=(8, 8))

# Original vector
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='b', label='v')

# Scaled vectors
plt.quiver(0, 0, scaled_v1[0], scaled_v1[1], angles='xy', scale_units='xy', scale=1, color='g', label='2v')
plt.quiver(0, 0, scaled_v2[0], scaled_v2[1], angles='xy', scale_units='xy', scale=1, color='r', label='-v')

plt.xlim(-3, 5)
plt.ylim(-3, 5)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Span of a Vector')
plt.axhline(0, color='gray', linewidth=0.8)
plt.axvline(0, color='gray', linewidth=0.8)
plt.grid(True)
plt.legend()
plt.show()
```

In this illustration, we have defined a vector **v (blue)**. We also calculate two scaled vectors:
- **2v (green)** and **v (red)**.

Observations:

- The original vector v (blue) serves as the starting point.
- The vectors 2v (green) and v (red) are obtained by scaling v using scalar multiples.
- The **span of the vector v includes all points that can be reached by scaling and adding multiples of v**, as represented by the green and red arrows.

## **Matrix**

**Defining matrix in contrast to a vector.**

A matrix is a **rectangular array of numbers, symbols, or expressions, arranged in rows and columns**. While matrices share some similarities with vectors, there are some key **differences**:

- **Vectors are one-dimensional arrays**, while **matrices are two-dimensional arrays**. Vectors have only one row or column, matrices have multiple rows and columns.
- The **entries in a vector are scalars**. The **entries in a matrix can be scalars, variables, or even expressions**.
- **Vectors represent a single linear entity**. **Matrices are used to represent systems of linear equations, transformations, and more**. Multiple vectors can be combined into a matrix.
- Mathematical operations defined for vectors (like addition) are not valid for matrices. Matrices have their own set of valid operations like matrix multiplication.

**So in summary, a matrix is a more general mathematical object than a vector. While vectors are special cases of matrices, matrices allow representing more complex linear algebraic relationships.**

### **Vectors vs Matrices**

Here is a table summarizing the key similarities and differences between vectors and matrices:

<div align="center">

Similarity / Difference | Vector | Matrix
-|-|-
Dimensionality | 1D array | 2D array
Size | Only has one column or row | Has multiple rows and columns
Entries | Usually numerical elements | Can be numbers, variables, expressions
Linearity | Forms linear combinations | Also exhibits linearity
Addition | Component-wise addition | Defined row-wise and column-wise
Multiplication | Dot product, scalar multiplication | Matrix product, scalar multiplication
Geometry | Magnitude and direction | Encodes coordinate transformations
Bases | Can represent basis vector | Columns can represent basis vectors
Applications | Spanning spaces | Transformations, data analysis, optimization

</div>

In summary:
- Vectors and matrices share linearity and some algebraic operations
- Matrices extend capability of vectors by adding dimensions, multiple components, and coordinate transformations
- Matrices enable more advanced linear algebra concepts and techniques

### **Matrix representation in NumPy**

In the NumPy library, matrices are represented using the
- numpy.array class, which is a multi-dimensional array.

While NumPy does provide a specialized
- numpy.matrix class,

**it is recommended to use the more general numpy.array for matrices as well, as it offers greater flexibility and compatibility with other NumPy functions.**

```python
import numpy as np

# Creating a 2x2 matrix
matrix_2x2 = np.array([[1, 2],
                       [3, 4]])

# Creating a 3x3 matrix
matrix_3x3 = np.array([[5, 6, 7],
                       [8, 9, 10],
                       [11, 12, 13]])

# Creating a 2x3 matrix
matrix_2x3 = np.array([[14, 15, 16],
                       [17, 18, 19]])

# Creating a 1x4 matrix (row vector)
row_vector = np.array([[20, 21, 22, 23]])

row_vector_2 = np.array([20, 21, 22, 23])

# Creating a 4x1 matrix (column vector)
column_vector = np.array([[24],
                          [25],
                          [26],
                          [27]])

print("2x2 Matrix:")
print(matrix_2x2, matrix_2x2.shape)

print("\n3x3 Matrix:")
print(matrix_3x3, matrix_3x3.shape)

print("\n2x3 Matrix:")
print(matrix_2x3, matrix_2x3.shape)

print("\nRow Vector:")
print(row_vector, row_vector.shape)

print("\nRow Vector 2:")
print(row_vector_2, row_vector_2.shape)

print("\nColumn Vector:")
print(column_vector, column_vector.shape)
```

### **Matrix Addition**

Matrix addition is an operation that combines two matrices of the **same dimensions element-wise**. Matrix addition follows the same rules as scalar addition.

Conditions for Matrix Addition:

- The matrices being added must have the same dimensions (same number of rows and same number of columns).
- The order of addition does not matter, i.e., A + B = B + A.

**However, beware of numpy broadcasting**

**Example 1: Simple Addition**

```python
# Define two matrices A and B
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

# Perform matrix addition
result = A + B

print("Matrix A:")
print(A)

print("\nMatrix B:")
print(B)

print("\nMatrix A + B:")
print(result)
```

**Example 2: Broadcasting with Different Shapes (Not for Matrix Addition)**


```python
A = np.array([[1, 2],
              [3, 4]])

scalar = 2

result = A + scalar

print(result)
```

**Example 3: Broadcasting with Different Shapes (Not for Matrix Addition)**

```python
A = np.array([[1, 2],
              [3, 4]])

B = np.array([5, 6])

result = A + B
print(result)
```

The inverse of a matrix is another matrix that, when multiplied with the original matrix, results in the identity matrix. For a square matrix \(A\), if there exists a matrix \(A^{-1}\) such that \(AA^{-1} = A^{-1}A = I\), where \(I\) is the identity matrix, then \(A^{-1}\) is the inverse of \(A\).

The importance of the inverse of a matrix lies in its role in solving linear systems of equations and various mathematical operations. Here are a few key points:

1. **Solving Linear Systems:**
   - Given a system of linear equations \(Ax = b\), where \(A\) is the coefficient matrix, \(x\) is the vector of variables, and \(b\) is the constant vector, you can find the solution \(x\) by multiplying both sides by the inverse of \(A\): \(x = A^{-1}b\).
   - This method is computationally efficient compared to other methods, especially when solving similar systems with different constant vectors.

2. **Matrix Division:**
   - In linear algebra, division by a matrix is not directly defined. However, matrix inversion serves a similar purpose. Instead of dividing by a matrix \(B\), you can multiply by the inverse of \(B\): \(A/B = AB^{-1}\).

3. **Numerical Stability:**
   - In numerical computations, solving linear systems using the inverse can be more stable than directly solving the system using Gaussian elimination or other methods, especially when dealing with ill-conditioned matrices.

4. **Coordinate Transformations:**
   - In geometry and computer graphics, the inverse of a transformation matrix can be used to undo a particular transformation, allowing for the restoration of original coordinates.

It's important to note that not all matrices have inverses. A matrix must be square (having the same number of rows and columns) and must be non-singular (its determinant must be non-zero) for it to have an inverse. If a matrix is singular, it is said to be "non-invertible."

Key points about the transpose and its significance:

1. **Symmetric Matrices:**
   - A square matrix is symmetric if its transpose is equal to itself (\(A = A^T\)). Symmetric matrices have special properties and are often encountered in various mathematical and scientific applications.

2. **Matrix Operations:**
   - Transposition is useful in various matrix operations. For example, the product of a matrix \(A\) and its transpose (\(AA^T\) or \(A^TA\)) results in a symmetric matrix.

3. **Vectors as Matrices:**
   - For a column vector, its transpose becomes a row vector, and vice versa. Transposition is used to switch between row and column vector representations.

4. **Orthogonal Matrices:**
   - Orthogonal matrices are square matrices whose transpose is also their inverse (\(A^TA = AA^T = I\)). They play a role in linear transformations and rotations.

5. **Efficient Memory Access:**
   - In programming and numerical computations, arranging data in row-major or column-major order in memory can affect performance. Transposing a matrix may be necessary for efficient memory access in certain situations.

6. **Linear Algebra Operations:**
   - Transposition is fundamental in linear algebra operations, such as solving linear systems, computing determinants, and diagonalizing matrices.

7. **Coordinate Transformations:**
   - In computer graphics and geometry, transposing transformation matrices is used to switch between different coordinate systems.

Overall, the transpose is a fundamental operation in linear algebra, offering insights into the structure of matrices and facilitating various mathematical and computational tasks.
