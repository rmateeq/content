### What is LA?

- Widely applied in various fields including physics, engineering, and machine learning.

### **Why Linear Algebra for Machine Learning or Deep Learning?**

Let's illustrate the role with examples:

#### **Foundation for working with high-dimensional data and transformations:**
Consider a dataset with three features: $x_1$, $x_2$, and $x_3$. The data for three data points could be represented as a matrix:

$ X = \begin{bmatrix} 1 & 2 & 3 \\ 4 & 5 & 6 \\ 7 & 8 & 9 \end{bmatrix} $

Here, each row represents a data point, and each column represents a feature. Linear algebra provides the foundation for handling and transforming such high-dimensional data.

#### **Essential for understanding and implementing machine learning algorithms:**
In linear regression, a fundamental machine learning algorithm, we have a linear relationship between features and a target variable. The relationship is expressed as a linear equation:

$ y\hat = b_0 + b_1x_1 + b_2x_2 + \ldots + b_nx_n $<br>
$y$

Here, $b_0, b_1, b_2, \ldots, b_n$ are coefficients, and $x_1, x_2, \ldots, x_n$ are features. The understanding of linear algebra is crucial for solving the system of linear equations involved in finding these coefficients.

#### **Many algorithms involve matrix operations, such as data preprocessing and regression:**
Principal Component Analysis (PCA), commonly used for dimensionality reduction, involves matrix operations. The covariance matrix of the data is computed, and eigenvectors are extracted. Matrix operations play a crucial role in these computations.

#### **Neural networks in deep learning involve matrix representations and operations:**
In deep learning, neural networks can be represented using matrices. The input layer, weights, and activations are all represented as matrices. The forward and backward passes involve matrix multiplications and other linear algebra operations.

For example, in a simple feedforward neural network, the output of a layer can be computed as:

$ \text{Output} = \text{Activation}(\text{Weight} \times \text{Input} + \text{Bias}) $

Here, the weight matrix, input matrix, and bias vector are fundamental components that rely on linear algebra operations.

### **What are the Building Blocks of Linear Algebra?**

- **Vectors:** Quantities with magnitude and direction; building blocks of matrices.
    - **Example:** Consider a 2D vector representing a position in space.
  $ \mathbf{v} = \begin{bmatrix} 3 \\ 4 \end{bmatrix} $
    - This vector has a magnitude of 5 units ($\|\mathbf{v}\| = 5$) and points in the direction of a right-angled triangle's hypotenuse.
- **Matrices:** Rectangular arrays of numbers; used for linear transformations and data representation.
  - **Example:** A matrix representing a 2x3 grid of pixel intensities in an image.
  - $ A = \begin{bmatrix} 255 & 0 & 128 \\ 64 & 192 & 31 \end{bmatrix} $
  - Here, each element of the matrix represents the intensity of a pixel in the image.

- **Scalars:** Single numeric values; opposites of vectors or matrices.
  - **Example:** A scalar representing the temperature in degrees Celsius.
  $ \text{Temperature} = 25 $
  - Unlike vectors or matrices, a scalar is a single numeric value.
- **Linear Transformations:** Operations that map vectors to vectors, preserving linear relationships.
  - **Example:** Scaling by a factor of 2 in 2D space.
  - $ \mathbf{T}(\mathbf{v}) = \begin{bmatrix} 2 & 0 \\ 0 & 2 \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix} = \begin{bmatrix} 2x \\ 2y \end{bmatrix} $
  - Linear transformations preserve the structure of vectors and can include operations like scaling, and rotation.
- **Linear Equations:** Equations involving vectors and matrices, often represented using matrix notation.
  - Already examplified.

### **An Example**
Let's consider an example of a 2-dimensional vector as an ordered pair in both mathematics and computer science.

### Mathematics (Linear Algebra):
In mathematics, a 2-dimensional vector is typically represented as an ordered pair (x, y). For example:

$ \mathbf{v} = \begin{bmatrix} 3 \\ -1 \end{bmatrix} $

Here, the vector $\mathbf{v}$ is represented as an ordered pair (3, -1). The first component (3) represents the magnitude along the x-axis, and the second component (-1) represents the magnitude along the y-axis. Geometrically, you can think of this vector as an arrow in a 2D space pointing from the origin (0,0) to the point (3, -1).

### Computer Science:
In computer science, vectors are often used to represent data in the form of arrays or lists. For a 2-dimensional vector, it might be represented as an ordered pair [x, y] in programming languages. For example, in Python:

```python
# Representing a 2D vector in Python
vector_in_python = [3, -1]
```

Here, the list `[3, -1]` serves as a representation of the 2-dimensional vector. The first element (3) corresponds to the x-coordinate, and the second element (-1) corresponds to the y-coordinate.

So, in summary:

- **Mathematics (Linear Algebra):**
  - **Vector Representation:** $\mathbf{v} = \begin{bmatrix} 3 \\ -1 \end{bmatrix}$
  - **Ordered Pair:** (3, -1)

- **Computer Science:**
  - **Vector Representation:** $[3, -1]$
  - **Ordered Pair:** (3, -1)

Both representations convey the same underlying concept of a 2-dimensional vector, but the context and usage may differ between the mathematical and computer science perspectives.

### **Calculating magnitude and angle of a vector**

To calculate the magnitude and direction of a 2D vector represented as $[2, 1]$, you can use the following formulas:

### Magnitude (or Length) of a 2D Vector:
The magnitude $(|\mathbf{v}|$) of a 2D vector $\mathbf{v} = [x, y]$ is calculated using the Pythagorean theorem:

$ |\mathbf{v}| = \sqrt{x^2 + y^2} $

For the vector $\mathbf{v} = [2, 1]$:

$ |\mathbf{v}| = \sqrt{2^2 + 1^2} = \sqrt{4 + 1} = \sqrt{5} $

So, the magnitude of the vector $[2, 1]$ is $\sqrt{5}$.

### Direction (or Angle) of a 2D Vector:
The direction (or angle) of a 2D vector $\mathbf{v} = [x, y]$ can be calculated using the arctangent function:

$ \text{Direction}(\mathbf{v}) = \text{atan2}(y, x) $

In many programming languages, the `atan2` function is available to handle the signs of $x$ and $y$ correctly. In Python, for example:

````python
import math

# Vector [2, 1]
x, y = 2, 1

# Calculate direction in radians
direction_radians = math.atan2(y, x)

# Convert radians to degrees
direction_degrees = math.degrees(direction_radians)

print(f"Direction (in radians): {direction_radians}")
print(f"Direction (in degrees): {direction_degrees}")
````

For the vector $[2, 1]$, the direction in radians and degrees will be calculated.

In this case, the result will be approximately:<br>
$ \text{Direction (in radians)} \approx 0.4636 \, \text{radians} $
$ \text{Direction (in degrees)} \approx 26.57 \, \text{degrees} $

So, for the vector $[2, 1]$, the magnitude is $\sqrt{5}$ and the direction is approximately $26.57$ degrees.

### Let's visualize it!

````python
# Define the vector components
x = 2
y = 1

# Calculate the magnitude and direction
magnitude = np.sqrt(x**2 + y**2)
direction_rad = np.arctan2(y, x)
direction_deg = np.degrees(direction_rad)

# Create a plot
plt.figure(figsize=(8, 8))

# Vector components
plt.quiver(0, 0, x, 0, angles='xy', scale_units='xy', scale=1, color='r', label='x-component')
plt.quiver(x, 0, 0, y, angles='xy', scale_units='xy', scale=1, color='g', label='y-component')

# Vector
plt.quiver(0, 0, x, y, angles='xy', scale_units='xy', scale=1, color='b', label='Vector')
plt.text(x + 0.2, y + 0.2, f'({x}, {y})', fontsize=12, color='b')

# Magnitude and Direction
plt.annotate('', xy=(x, y), xytext=(0, 0), arrowprops=dict(arrowstyle='->', color='m'), color='m')
plt.text(x / 2, -0.5, f'x = {x}', fontsize=12, color='r')
plt.text(-0.5, y / 2, f'y = {y}', fontsize=12, color='g')
plt.text(-0.2, -0.5, f'Magnitude: {magnitude:.2f}', fontsize=12, color='m')
plt.text(-0.2, -0.7, f'Direction: {direction_deg:.2f}', fontsize=12, color='m')

plt.xlim(-1, 3)
plt.ylim(-1, 3)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Vector Components, Magnitude, and Direction')
plt.axhline(0, color='gray', linewidth=0.8)
plt.axvline(0, color='gray', linewidth=0.8)
plt.grid(True)
plt.legend()
plt.show()
````

### **Operations: Addition and scalar multiplication**

Vector addition and scalar multiplication are two of the most fundamental operations defined on vectors in linear algebra.

**Vector Addition:**
- Adding two vectors involves adding their corresponding elements.
- Geometrically, this corresponds to placing vectors head to tail and finding the resulting vector from the origin to the endpoint.
- For example, adding [1, 2] and [3, 4] gives [1+3, 2+4] = [4, 6].

**Scalar Multiplication:**  
- Multiplying a vector by a scalar scales the vector by that amount.
- Geometrically, the vector's length is stretched or shrunk while the direction remains the same.
- For example, multiplying [1, 2] by 3 gives [3\*1, 3\*2] = [3, 6].

These operations satisfy key properties like associativity, commutativity, distributivity, etc. They form the **basis for manipulating vectors** in linear algebra and define the vector space structure.

### Let's visualize multiplication!

````python
# Define the vector
vector = np.array([2, 1])

# Define scaling factors
positive_scale = 2
negative_scale = -0.5

# Calculate scaled vectors
scaled_positive = positive_scale * vector
scaled_negative = negative_scale * vector

# Create a plot
plt.figure(figsize=(10, 5))

# Original vector
plt.subplot(1, 3, 1)
plt.axhline(0, color='gray', linewidth=0.8)
plt.axvline(0, color='gray', linewidth=0.8)
plt.quiver(0, 0, vector[0], vector[1], angles='xy', scale_units='xy', scale=1, color='b')
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Original Vector [2 1]')
plt.grid(True)

# Positive scaling
plt.subplot(1, 3, 2)
plt.axhline(0, color='gray', linewidth=0.8)
plt.axvline(0, color='gray', linewidth=0.8)
plt.quiver(0, 0, scaled_positive[0], scaled_positive[1], angles='xy', scale_units='xy', scale=1, color='g')
plt.xlim(-3, 5)
plt.ylim(-3, 3)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Positive Scaling (2)')
plt.grid(True)

# Negative scaling
plt.subplot(1, 3, 3)
plt.axhline(0, color='gray', linewidth=0.8)
plt.axvline(0, color='gray', linewidth=0.8)
plt.quiver(0, 0, scaled_negative[0], scaled_negative[1], angles='xy', scale_units='xy', scale=1, color='r')
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Negative Scaling (-0.5)')
plt.grid(True)

plt.tight_layout()
plt.show()
````

So scaling **does not change the origin of vector and changes the length and/or direction!**

### Let's visualize additon!

````python
# Define the vectors
vector1 = np.array([2, 1])
vector2 = np.array([1, 2])

# Calculate the sum vector
sum_vector = vector1 + vector2

# Create a plot
plt.figure(figsize=(10, 5))

# Vector 1
plt.subplot(1, 3, 1)
plt.axhline(0, color='gray', linewidth=0.8)
plt.axvline(0, color='gray', linewidth=0.8)
plt.quiver(0, 0, vector1[0], vector1[1], angles='xy', scale_units='xy', scale=1, color='b')
plt.xlim(-1, 4)
plt.ylim(-1, 4)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Vector [2 1]')
plt.grid(True)

# Vector 2
plt.subplot(1, 3, 2)
plt.axhline(0, color='gray', linewidth=0.8)
plt.axvline(0, color='gray', linewidth=0.8)
plt.quiver(0, 0, vector2[0], vector2[1], angles='xy', scale_units='xy', scale=1, color='g')
plt.xlim(-1, 4)
plt.ylim(-1, 4)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Vector [1 2]')
plt.grid(True)

# Sum vector
plt.subplot(1, 3, 3)
plt.axhline(0, color='gray', linewidth=0.8)
plt.axvline(0, color='gray', linewidth=0.8)
plt.quiver(0, 0, vector1[0], vector1[1], angles='xy', scale_units='xy', scale=1, color='b')
plt.quiver(vector1[0], vector1[1], vector2[0], vector2[1], angles='xy', scale_units='xy', scale=1, color='g')
plt.quiver(0, 0, sum_vector[0], sum_vector[1], angles='xy', scale_units='xy', scale=1, color='r')
plt.xlim(-1, 4)
plt.ylim(-1, 4)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Vector Sum [3 3]')
plt.grid(True)

plt.tight_layout()
plt.show()
````

**Summing two vectors is as if you place the second vector at the tip of first vector and then draw a new vector from origin to the tip of second vector!**

### **Eplaninig the origin! Why (0,0)?**

The origin (0, 0) is a fundamental point in a Cartesian coordinate system and serves as the reference point for defining all other points. In linear algebra, vectors are often represented as directed line segments from the origin to a specific point in space. This choice of origin is a convention that simplifies mathematical calculations and geometric interpretations.

Using the origin as the reference point provides several benefits:

- Simplicity: The origin simplifies mathematical operations involving vectors, such as addition, subtraction, and scaling.

- Consistency: By choosing a fixed origin, vector operations are consistent and do not depend on the specific application.

- Geometric Interpretation: The origin represents the "starting point" for vectors, making it easier to visualize and understand vector relationships and transformations.

- Coordinate System: The origin defines the point where the x-axis and y-axis intersect, establishing the coordinate system for locating other points in space.

Overall, the choice of the origin (0, 0) as the reference point is a foundational convention that underlies many concepts in linear algebra and facilitates clear and consistent mathematical reasoning.
